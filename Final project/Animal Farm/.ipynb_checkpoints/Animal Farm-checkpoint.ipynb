Steps
1. Read the file
2. Each line in the file will be an item of the list
3. Clean the list and its items
4. Figure out where does the book starts and where it finishes
5. Separate the whole novel by words
6. Discover the most used pronouns
7. Discover the main characters in the novel

```python
#Step 1 : Read the file.
fullbook = open("animals.txt", "r",encoding="utf-8")
print(fullbook)
```

```python
#step 2 Create the list
read = fullbook.readlines()
print(type(read))
```

```python
#step 3: Clean the list and its items

for item in read:
    if item == "\n":
        read.remove(item)
print(read)


#fullbook = open("book_1984.txt", "r",encoding="utf-8")
#book1984 = []
#for readlines in fullbook:
 #   book1984.append(readlines.strip("\n"))
#for readlines in fullbook:
  #  if readlines == " '' ":
   #     fullbook.remove(readlines)
#print(book1984)
```

```python
newRead = []
for item in read:
    newRead.append(item.strip("\n"))
print(newRead)
```

```python
#step 4
#Figure out where does the book starts and where it finishes
start = newRead.index("Chapter 1")
print(start)

end = newRead.index("THE END")
print(end)
```

```python
newRead = newRead[start:end]
print(newRead)
```

```python
#step 5: seprate word
allLines = " ".join(newRead)
print(allLines)
```

```python
allwords = allLines.split()
print(allwords)
```

```python
#step 6:Find the pronouns
pronouns = ["I", "you", "You", "He", "he", "She", "she", "It", "it", "We", "we", "They", "they"]
used_pronouns = []
for words in allwords:
    if words in pronouns:
        used_pronouns.append(words.lower())
print(used_pronouns)
```

```python
used_pronouns = set(used_pronouns)
list_pronouns =list(used_pronouns)
print(list_pronouns)
```

```python
#counter
n = 0
while n < len(list_pronouns):
    print(f"'{list_pronouns[n].capitalize()}' appears {used_pronouns.count(list_pronouns[n])} times.")
    n += 1
```

```python
#step 7: Find out the Protagonist
capitalized_words = []
for word in allwords:
    if word.istitle() == True:
        capitalized_words.append(word)
print(capitalized_words)   
```

```python
list_capitalized = list(set(capitalized_words))
stopwords = ["I", "He", "She", "You", "We", "They", "And", "A", "The", "There", "It", "That", "But", "In", "On"]
n = 0
while n < len(list_capitalized):
    if list_capitalized[n] in stopwords:
        None
    else:
        if capitalized_words.count(list_capitalized[n]) > 50:
            print(f"The word '{list_capitalized[n]}' appears {capitalized_words.count(list_capitalized[n])} times.")
    n += 1
```
